```python
from neo4j import GraphDatabase

# Define the connection details
uri = "neo4j://localhost:7687"
username = "neo4j"
password = "password"

# Connect to the Neo4j database

driver = GraphDatabase.driver(uri, auth=(username, password))
```

```python
def delete_all_relationships(session):
    # Run a Cypher query to delete all relationships
    query = "MATCH ()-[r]-() DELETE r"
    session.run(query)

# Create a session and execute the deletion
with driver.session() as session:
    delete_all_relationships(session)
```

```python
# Create a session
with driver.session() as session:   #Cypher query
    # Delete all nodes and relationships
    session.run("MATCH (n) DETACH DELETE n")
```

```python
def delete_nodes_in_batches(session, batch_size=1000):
    query = "MATCH (n) WITH n LIMIT $batchSize DETACH DELETE n"
    session.run(query, batchSize=batch_size)

with driver.session() as session:
    delete_nodes_in_batches(session)

driver.close()
```

```python
def delete_all_nodes_and_relationships(session):
    # Run a Cypher query to delete all nodes and relationships
    query = "MATCH (n) DETACH DELETE n"
    session.run(query)
    
# Create a session and execute the deletion
with driver.session() as session:
    delete_all_nodes_and_relationships(session)
```

```python
def create_relationship_between_nodes(node1_label, node2_label, relationship_type):
    with driver.session() as session:
        query = (
        f"MATCH (n1:{node1_label}), (n2:{node2_label}) "   
        f"WHERE n1.id = n2.id "
       f"CREATE (n1)-[:{relationship_type}]->(n2)"
        )
        session.run(query)


# Create relationships between nodes  Event
create_relationship_between_nodes('Patient_Information','Event_Information',  "PARTICIPATED_IN")

# Create relationships between nodes  enforcement_{i}
create_relationship_between_nodes('Recall_Details','Location_Information',  "AFFECTED_LOCATIONS")
create_relationship_between_nodes('Recall_Details','Drug_Information',  "RELATED_DRUGS")
create_relationship_between_nodes('Recall_Details','OpenFDA_Information',  "FDA_REPORTS")

# Create relationships between nodes  ndc_{i}
create_relationship_between_nodes('Drug_Information','Active_Ingredients',  "CONTAINS_INGREDIENT")
create_relationship_between_nodes('Drug_Information','Marketing_and_Product_Lifecycle',  "RELATED_LIFECYCLE_INFO")
create_relationship_between_nodes('Drug_Information','OpenFDA_Information',  "FDA_REPORTED")


# Create relationships between nodes  label_{i}_{j}
create_relationship_between_nodes('Drug_Information','Abuse_and_Overdosage',  "CONTAINS_ABUSE_INFO")
create_relationship_between_nodes('Drug_Information','Active_Ingredients',  "HAS_ACTIVE_INGREDIENT")
create_relationship_between_nodes('Drug_Information','Clinical_Pharmacology',  "PHARMACOLOGY_REPORTED")
create_relationship_between_nodes('Drug_Information','Contraindications',  "CONTAINS_CONTRAINDICATION")
create_relationship_between_nodes('Drug_Information','Safety_and_Adverse_Effects',  "HAS_ADVERSE_EFFECTS_INFO")
create_relationship_between_nodes('Drug_Information','Warnings_and_Precautions',  "FDA_WARNINGS")
create_relationship_between_nodes('Drug_Information','Use_in_Specific_Populations',  "CONTAINS_POPULATION_INFO")
create_relationship_between_nodes('Drug_Information','Supply_storage_and_handling',  "LIFECYCLE_STORAGE_INFO")
create_relationship_between_nodes('Drug_Information','Nonclinical_toxicology',  "TOXICOLOGY_REPORTED")
create_relationship_between_nodes('Drug_Information','Documentation_and_Additional_Information',  "CONTAINS_DOCUMENTATION")
create_relationship_between_nodes('Drug_Information','OpenFDA_Information',  "RELATED_LIFECYCLE_INFO")
```
