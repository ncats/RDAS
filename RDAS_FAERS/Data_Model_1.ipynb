```python
# Define node types
node_types = [
    "Event", 
    "Event_primarysource",
    "Event_sender",
    "Event_patient",
    "Event_patient_PatientDeath",
    "Event_patient_drug", 
    "Event_patient_drug_open_fda", 
    "Drug", 
    "Drug_openfda", 
    "Drug_Recall", 
    "Drug_Recall_openfda", 
    "drug_ndc", 
    "drug_ndc_openfda"
]

# Create nodes
with driver.session() as session:
    for node_type in node_types:
        query = f"CREATE (:`{node_type}`)"
        session.run(query)
```

```python
def create_event_node(properties):
    with driver.session() as session:
        query = (
            "CREATE (e:Event {"
            "id: $id, "
            "safetyreportid: $safetyreportid, "
            "transmissiondateformat: $transmissiondateformat, "
            "transmissiondate: $transmissiondate, "
            "serious: $serious, "
            "seriousnessdeath: $seriousnessdeath, "
            "receivedateformat: $receivedateformat, "
            "receivedate: $receivedate, "
            "receiptdateformat: $receiptdateformat, "
            "receiptdate: $receiptdate, "
            "fulfillexpeditecriteria: $fulfillexpeditecriteria, "
            "companynumb: $companynumb ,"
            "receiver: $receiver"
            "})"
        )

        session.run(query, **properties)

def create_event_primarysource_node(properties):
    with driver.session() as session:
        query = (
            "CREATE (ep:Event_primarysource {"
            "id: $id, "
            "reportercountry: $reportercountry, "
            "qualification: $qualification"
            "})"
        )

        session.run(query, **properties)

def create_relationship_between_event_and_primarysource(event_id, primarysource_id):
    with driver.session() as session:
        query = (
            "MATCH (e:Event), (ep:Event_primarysource) "
            "WHERE e.id = $event_id AND ep.id = $primarysource_id "
            "CREATE (e)-[r:PRIMARY_SOURCE_OF_EVENT]->(ep) "
            "RETURN r"
        )

        session.run(query, event_id=event_id, primarysource_id=primarysource_id)

# Assuming you have unique identifiers for Event and Event_primarysource nodes.

for i in range(200):
    event_data = json_data["results"][i]
    
    # Create Event node
    event_properties = {
        "id": f"event_{i}",
        "safetyreportid": event_data.get('safetyreportid'),
        "transmissiondateformat": event_data.get('transmissiondateformat'),
        "transmissiondate": event_data.get('transmissiondate'),
        "serious": event_data.get('serious'),
        "seriousnessdeath": event_data.get('seriousnessdeath'),
        "receivedateformat": event_data.get('receivedateformat'),
        "receivedate": event_data.get('receivedate'),
        "receiptdateformat": event_data.get('receiptdateformat'),
        "receiptdate": event_data.get('receiptdate'),
        "fulfillexpeditecriteria": event_data.get('fulfillexpeditecriteria'),
        "companynumb": event_data.get('companynumb'),
        "receiver": event_data.get('receiver')
    }
    create_event_node(event_properties)
    
    primarysource_data = event_data.get('primarysource', {})
    
    # Ensure 'primarysource' exists and is a dictionary
    if not isinstance(primarysource_data, dict):
        primarysource_data = {}
    
    # Add missing properties with default value 'none'
    primarysource_data.setdefault('reportercountry', 'none')
    primarysource_data.setdefault('qualification', 'none')
    
    # Update the json_data
    json_data["results"][i]['primarysource'] = primarysource_data
    
    # Create Event_primarysource node
    primarysource_properties = {
        "id": f"primarysource_{i}",
        "reportercountry": primarysource_data.get('reportercountry'),
        "qualification": primarysource_data.get('qualification')
    }
    create_event_primarysource_node(primarysource_properties)
    
    # Create relationship between Event and Event_primarysource nodes
    create_relationship_between_event_and_primarysource(
        event_properties['id'],
        primarysource_properties['id']
    )
```

```python
def create_sender_node(sender_id, senderorganization):
    with driver.session() as session:
        query = (
            "CREATE (s:Event_sender {id: $sender_id, senderorganization: $senderorganization})"
        )

        session.run(query, sender_id=sender_id, senderorganization=senderorganization)


def create_patient_node(patient_id, properties):
    with driver.session() as session:
        query = (
            "CREATE (p:Event_patient {"
            "id: $patient_id, "
            "patientonsetage: $patientonsetage, "
            "patientonsetageunit: $patientonsetageunit, "
            "patientweight: $patientweight, "
            "patientsex: $patientsex "
            "})"
        )

        session.run(query, patient_id=patient_id, **properties)

def create_patientdeath_node(patient_death_id, properties):
    with driver.session() as session:
        query = (
            "CREATE (pd:Event_patient_PatientDeath {"
            "id: $patient_death_id, "
            "patientdeathdateformat: $patientdeathdateformat, "
            "patientdeathdate: $patientdeathdate"
            "})"
        )

        session.run(query, patient_death_id=patient_death_id, **properties)

# Assuming you have unique identifiers for Sender, Patient, and PatientDeath nodes.

for i in range(200):
    sender_data = json_data["results"][i].get('sender', {})
    if not isinstance(sender_data, dict):
        sender_data = {}
    sender_organization = sender_data.get('senderorganization', 'None')
    create_sender_node(i, sender_organization)

    patient_data = json_data["results"][i].get('patient', {})
    if not isinstance(patient_data, dict):
        patient_data = {}
    patient_properties = {
        "patientonsetage": patient_data.get('patientonsetage'),
        "patientonsetageunit": patient_data.get('patientonsetageunit'),
        "patientweight": patient_data.get('patientweight'),
        "patientsex": patient_data.get('patientsex')
    }
    create_patient_node(i, patient_properties)

    patient_death_data =  patient_data.get('patientdeath', {})
    if not isinstance(patient_death_data, dict):
        patient_death_data = {}
    patient_death_properties = {
        "patientdeathdateformat": patient_death_data.get('patientdeathdateformat'),
        "patientdeathdate": patient_death_data.get('patientdeathdate')
    }
    create_patientdeath_node(i, patient_death_properties)
```

```python
def create_node(tx, label, key):
    tx.run(f"CREATE (n:{label} {{name: $name}})", name=key)

def create_relationship(tx, parent, child):
    tx.run("MATCH (p {name: $parent}), (c {name: $child}) "
           "CREATE (p)-[:HAS]->(c)", parent=parent, child=child)
# Define the nested keys
keys = [
    "meta",
    "disclaimer",
    "terms",
    "license",
    "last_updated",
    "results",
    "results.skip",
    "results.limit",
    "results.total"
]

with driver.session() as session:
    for key in keys:
        labels = key.split('.')
        for i, label in enumerate(labels):
            session.write_transaction(create_node, label, label)
            if i > 0:
                parent_label = labels[i-1]
                session.write_transaction(create_relationship, parent_label, label)
```
