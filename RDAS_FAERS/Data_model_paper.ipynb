```python
#deleteing the previous database
def delete_nodes_in_batches(batch_size=1000):
    with GraphDatabase.driver(uri, auth=(username, password)) as driver:
        with driver.session() as session:
            while True:
                # Delete nodes in batches to avoid hitting memory limits
                query = f"MATCH (n) WITH n LIMIT {batch_size} DETACH DELETE n RETURN count(n) as deleted_nodes"
                result = session.run(query)
                deleted_nodes = result.single()["deleted_nodes"]
                
                if deleted_nodes == 0:
                    break
                print(f"Deleted {deleted_nodes} nodes in this batch.")

# Run the batch deletion
delete_nodes_in_batches(batch_size=10000)
```

```python
def print_all_nodes():
    with driver.session() as session:
        result = session.run("MATCH (n) RETURN n")
        for record in result:
            print(record["n"])

# Call the function
print_all_nodes()

# Close the driver connection when done
driver.close()
```

```python
def print_node_categories():
    with driver.session() as session:
        # Get distinct labels (categories) of nodes
        result = session.run("MATCH (n) RETURN DISTINCT labels(n) AS labels")
        
        print("Node categories (labels):")
        for record in result:
            print(record["labels"])

# Call the function
print_node_categories()

# Close the driver connection when done
driver.close()
```

```python
def print_relationship_types():
    with driver.session() as session:
        # Get distinct relationship types
        result = session.run("MATCH ()-[r]->() RETURN DISTINCT type(r) AS relationship_type")
        
        print("Relationship types (edge labels):")
        for record in result:
            print(record["relationship_type"])

# Call the function
print_relationship_types()

# Close the driver connection when done
driver.close()
```
